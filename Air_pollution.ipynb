```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\champ\Downloads\\3b01bcb8-0b14-4abf-b6f2-c1bfd384ba69.csv")
df
```

```python
top_states = df.groupby('state')['pollutant_avg'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_states.values, y=top_states. index, palette='rocket')
plt.title('Top 10 States by Average Pollution Level')
plt.xlabel('Average Pollutant Level' )
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
top_cities = df.groupby('city') ['pollutant_avg'].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(8, 5))
top_cities.plot(kind='bar', color='crimson')
plt.title('Top 5 Most Polluted Cities')
plt.ylabel('Average Pollution Level')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.tight_layout()
plt. show()
```

```python
top_states = df.groupby('state')['pollutant_avg'].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(7,7))
top_states.plot.pie(autopct='%1.1f%%', cmap='Set3', startangle=90)
plt.title('Top 5 States by Avg Pollution Share')
plt.ylabel('')
plt. show()
```

```python
pollutant = 'PM2.5'
top_states_pm25 = df[df['pollutant_id' ] == pollutant].groupby('state' ) ['pollutant_avg' ].mean() . sort_values(ascending=False)
print(top_states_pm25.head(10))
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['longitude' ], df['latitude' ], c=df['pollutant_avg'], cmap='coolwarm', alpha=0.6)
plt.colorbar(scatter, label='Average Pollutant Level')
plt.title('Geographic Distribution of Pollution')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt. tight_layout ()
plt. show()
```
